```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)
points_df
# find the euclidean distance
# bet two points (x2-x1)^2 + (y2-y1)^2
```

```python
def euclidean_dist(x1,x2,y1,y2):
    return ((x2-x1)^2) + ((y2-y1)^2)
```

```python
euclidean_dist(3,-3,4,4)
```

```python
def euclidean_dist(row):
    point_a =  row['1st point']
    point_b = row['2nd point']
    
    return ((point_b[0]-point_a[0])**2 + (point_b[1]-point_a[1])**2)**0.5
```

```python
points_df['euclidean_distance'] = points_df.apply(euclidean_dist,axis=1)
```
